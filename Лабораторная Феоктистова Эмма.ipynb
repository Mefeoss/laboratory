```python
# Берем текст
text = 'Мистер Хангертон, отец моей Глэдис, отличался невероятной \
бестактностью и был похож на распушившего перья неопрятного  какаду, \
правда, весьма добродушного, но занятого исключительно собственной особой. \
Если что-нибудь могло оттолкнуть меня от Глэдис, так только крайнее \
нежелание обзавестись глуповатым тестем. Мистер Хангертон желал Глэдис \
только счастья.'

words = text.replace(',', '').replace('.', '').split()
```

```python
def addPostfix(postfix, curLevel):
    # Префикс закончился (его постфикс пуст). Добавлять больше нечего.
    if len(postfix) == 0:
        return
    # Такого начала нет в словаре. Его надо добавить.
    if curLevel.get(postfix[0], None) == None:
        curLevel[postfix[0]] = [0, {}]
    # Пришел еще один такой префикс, увеличиваем частоту.
    curLevel[postfix[0]][0] += 1
    # Добавляем оставшуюся часть в соответствующий узел.
    addPostfix(postfix[1:], curLevel[postfix[0]][1])
```

```python
amount = int(input('Введите число: ')) # спрашиваем у пользователя цифру
a = amount
ngrams = []
    
for i in range(0, len(words) - amount + 1):
    ngrams.append(words[i:amount])
    amount += 1
    # составляем список n-грамм
```

```python
allNgrams= {}

for ngram in ngrams:
    addPostfix(ngram, allNgrams)
    # записываем значения в префиксное дерево
```

```python
print(allNgrams)
```

```python
from collections import Counter
count = Counter()


for ngram in ngrams:
    if len(ngram) == a:
        string = ' '.join(ngram)
        count[string] += 1 
```

```python
print(count) # подсчитываем частотность n-грамм
```

```python
print(count.most_common()) # выводим самые частотные n-граммы
```
